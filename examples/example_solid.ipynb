```python
from pyGmsh import pyGmsh
from pyGmsh import Numberer, FEMData
import numpy as np
import openseespy.opensees as ops
from pathlib import Path
```

```python
model_iges_path = Path(r'C:\Users\nmora\Github\pyGmsh\acad') / 'solid.iges'
assert model_iges_path.exists(), model_iges_path

m1 = pyGmsh(
    model_name='solid',
    verbose=True
)
m1.initialize()

m1.model.load_iges(
    file_path=model_iges_path,
    highest_dim_only=True
)

#m1.physical.add_curve(tags=[6, 7], name="UxUy")
#m1.physical.add_curve(tags=[2, 4], name="load")

m1.model.viewer()


(m1.mesh
     .set_size_sources(from_points=False)
     .generate(dim=3)
)

m1.mesh.viewer()

m1.mesh.renumber_mesh(method='rcm')

fem_data = m1.mesh.get_fem_data()
```

```python
cut_1 = m1.model.add_rectangle(
    x=-1,
    y=-1,
    z=0.5,
    dx=10,
    dy=10,
)

cut_2 = m1.model.add_rectangle(
    x=-1,
    y=-1,
    z=3.00,
    dx=10,
    dy=10,
)

m1.model.viewer()

m1.model.fragment(
    objects=[1],
    tools=[cut_1, cut_2],
    dim=3,
)

m1.model.viewer()
```
