## Loading CAD models

We will first start by loading a set of elements modeled using CAD tools and exported to .IGES format.
GMSH work well with both .STEP files and .IGES files.
You can choose your prefer CAD modeling software



```python
solid_CAD_file = Path(r'C:\Users\nmora\Desktop\gshTest\acad\SolidColumn')
solid_CAD_file = solid_CAD_file / 'SolidColumn_RevA.iges'

gmsh.initialize()
gmsh.model.add("SolidColumn")

# Load the IGES file using OpenCASCADE kernel (required for IGES/STEP files)
gmsh.model.occ.importShapes(str(solid_CAD_file))
gmsh.model.occ.synchronize()

print(f"Gmsh version: {gmsh.__version__}")
print(f"Loaded: {solid_CAD_file.name}")

# Launch the GUI to inspect the geometry
gmsh.fltk.run()

gmsh.finalize()
```

```python
# Two ways of using either with the with block: or by manually calling initialize() and finalize()

model = apeGmsh(model_name="SolidColumn", verbose=True)
with model:
    solid_CAD_file = Path(r'C:\Users\nmora\Desktop\gshTest\acad\SolidColumn') / 'SolidColumn_RevA.iges'
    model.load_geometry_iges(solid_CAD_file)
    
    #model.launch_gui()
    #model.plot_geometry()
    
    mapping, type_summary = model.inspect.get_geometry_info()
    print(type_summary)
    print(mapping)
    
    print(mapping['points'])
```
